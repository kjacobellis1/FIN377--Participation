```python
import pandas as pd
import os 
from requests_html import HTMLSession
import re
from time import sleep

session = HTMLSession()

###Section 1: get the urls to crawl


#if i dont have the list/data with URLS
    #1. acquire the data/urls to crawl
    #2. save that data/urls
#else
    #open that file
    
    
    
if not os.path.exists('inputs/'):
    os.makedirs('inputs/')
    
if not os.path.exists('inputs/sp500_wiki_urls.csv'):   

     #open page   
    url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
    session = HTMLSession()
    r=session.get(url)

    #grab the URLs
    table= r.html.find('#constituents')[0]
    table_rows=table.find('tr')
    colinks=[] #creating an empty frame to store this data in 
    for row in table_rows[1:]:
         colinks.append(list(row.find('td')[1].absolute_links)[0])

    #save it        
    if not os.path.exists('inputs/'):
        os.makedirs('inputs/')
    pd.DataFrame(colinks).to_csv('inputs/sp500_wiki_urls.csv',index=False,header=False)

else:
    colinks=pd.read_csv('inputs/sp500_wiki_urls.csv',names=['firm'])['firm'].to_list()

    
###Section 2: crawl the urls

for url in colinks[:15]: #row in the data structure with urls or info that leads to urls: 
                    #Q0: whats the url
    folder='wiki_html/'   #Q1: whats the folder to save the html
    if not os.path.exists('wiki_html/'):
        os.makedirs('wiki_html/')
        
        
    name = re.sub(r'\W+','',     url.split('/')[-1])
    
    filepath=folder + name + '.txt'
    
    #filepath= folder + '/' + re.sub(r'/W+','',url.split('/')[-1]) + '.txt' #Q2: whats the filename?
    
    
    if not os.path.exists(filepath): #<filename this page would get>
        try:
            r = session.get(url)
        except:
            print('sad face failure:', url)
        else:
            with open(filepath, 'w', encoding = 'utf8') as f:
                f.write(r.text)
            sleep(2) #be nice to server 
  
  
  #  Q1+Q2: = path/filename ill save it to
        
    #if not os.path.exists(<filename this page would get>): 
     #   try:
           # r = requests.get(<the url>)
     #   except:
            # log the error somehow
     #   else:
            # save the results, I typically save the RAW source
     #       sleep(3) # be nice to server
```
