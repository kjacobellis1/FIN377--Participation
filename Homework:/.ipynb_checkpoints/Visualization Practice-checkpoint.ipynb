## More Visualization Practice

```python
import pandas as pd
import numpy as np
import pandas_datareader as pdr # you might need to install this (see above)
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

# don't copy and paste these lists and number in several places 
# (which is what the original "simple program" did... that's bad programming!)
# instead, create a var - if we change it here, it changes everywhere
stocks = ['AAPL','MSFT','VZ']
start_yr = 2000

# load
stock_prices = pdr.get_data_yahoo(stocks, start=start_yr)
stock_prices = stock_prices.filter(like='Adj Close') # reduce to just columns with this in the name
stock_prices.columns = stocks # put their tickers as column names
stock_prices # print
```

```python
## Q1: Check that there are 3 missing observations of ret: the first return for each shouldn't exist.

stock_prices = stock_prices.stack().swaplevel().sort_index().reset_index()
stock_prices.columns=['Firm','Date','Adj Close'] #renames columns 
stock_prices['ret']=stock_prices.groupby('Firm')['Adj Close'].pct_change()

stock_prices['ret'].isnull().sum() #shows that there are 3 missing observations of ret  
```

```python
## Q2: Describe returns for each firm separately (i.e. make one table)...


stock_prices.groupby('Firm')['ret'].describe().transpose() #transpose flips it so the firms are the columns
                                                            # says: for each firm, describe the return variable 
```

```python
## Q3: Replicate and then improve each of the three figures below.

#figure 1: before
sns.distplot(stock_prices['ret'].dropna(),kde=False)
```

```python
#figure 1: after
sns.distplot(stock_prices.query('abs(ret)<.1')['ret'].dropna(),kde=False) #use the query to ignore the crazy outliers
```

```python
# figure 2: before 

sns.boxplot(stock_prices['ret'])
```

```python
# figure 2: after

sns.boxplot(stock_prices.query('abs(ret)<.1')['ret']) #query allows you to look more closely at the 25th and 75th percentile
```

```python
#figure 3: before


sns.lineplot(data=stock_prices.query('abs(ret)<0.2'),x='Date',y='ret', hue='Firm') #how to fix this? -- 3 separate graphs,
                                                            #making it weekly/monthly
```

```python
#figure 3: after
import matplotlib.pyplot as plt

plt.style.use('seaborn-darkgrid')


# Create a grid : initialize it
g = sns.FacetGrid(stock_prices, col='Firm', hue='Firm', col_wrap=3, height=4.0,margin_titles=True)

# Add the line over the area with the plot function
g = g.map(plt.plot, 'Date', 'ret')
 
# Fill the area with fill_between
g = g.map(plt.fill_between, 'Date', 'ret', alpha=0.9).set_titles("{col_name}")
g.set_xticklabels(rotation = 45)  
  

# Control the title of each facet
g = g.set_titles("{col_name}")


axes = g.axes.flatten()

for ax in axes:
    ax.set_xlabel("Fiscal Year",fontsize=12)
    ax.set_ylabel("Average Return",fontsize=12)
# Add a title for the whole plot
plt.subplots_adjust(top=0.72)
g = g.fig.suptitle('Average Stock Return For Three Firms',fontsize=22)


plt.show()
```
