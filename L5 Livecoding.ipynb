# Got - Baby Names

```python
#TO USE datadotworld PACKAGE:
#1. create account at data.world, then run the next two lines:
#2. in terminal/powershell: pip install datadotworld[pandas]
#
#   IF THIS DOESN'T WORK BC YOU GET AN ERROR ABOUT "CCHARDET", RUN:
#    conda install -c conda-forge cchardet
#   THEN RERUN: pip install datadotworld[pandas] 
#
#3. in terminal/powershell: dw configure          
#3a. copy in API token from data.world (get from settings > advanced)

import datadotworld as dw
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

baby_names = dw.load_dataset('nkrishnaswami/us-ssa-baby-names-national')
baby_names = baby_names.dataframes['names_ranks_counts']

```

```python

baby_names
baby_names.head(10)
baby_names.shape
```

```python
#look at number variables
baby_names.describe() #does it all make sense? any data issues?

#look at caregory variables
baby_names['sex'].value_counts() # this isnt 50/50!... but thats ok
baby_names['sex'].nunique() 
baby_names['year'].value_counts()
```

```python
#lets count Johns over time... 

somenames = baby_names.loc[baby_names['name'] == 'John']
combined = somenames.groupby(['name','year'])['count'].sum().reset_index() #split the entire data set up  #.sum sums every observation within the group --> ex: John 1990 7; John 1991 12 --> then combine
combined #reset index to turn year into a variable!!
sns.lineplot(data=combined, x='year', y='count') 
```

```python
# look at arya and sansa # just copy data set from above

somenames = baby_names.loc[baby_names['name'] == 'Arya']
combined = somenames.groupby(['name','year'])['count'].sum().reset_index() 
combined 
sns.lineplot(data=combined, x='year', y='count') 
```

```python
names_i_want = (baby_names['name'] == 'Sansa') | (baby_names['name'] == 'Arya')  # | means 'or' to pandas
names_i_want.sum() #only 6 years where there is a Sansa, additional 64 Aryas
somenames = baby_names.loc[names_i_want]
combined = somenames.groupby(['name','year'])['count'].sum().reset_index() 
combined 
sns.lineplot(data=combined, x='year', y='count', hue='name') 
```

```python
# QUERY!!
somenames = baby_names.query('name in ["Arya","Sansa"]') # query needs a string with true/false conditions
combined = somenames.groupby(['name','year'])['count'].sum().reset_index() 
combined 
sns.lineplot(data=combined, x='year', y='count', hue='name') 
```

```python
# i dont want "somenames"
(
    baby_names
    .query('name in ["Arya","Sansa"]') #get a subsample
    .groupby(['name','year'])['count'].sum().reset_index() #combine M-F
                                         # "pipe" in a function to apply to this object
                                        # syntax: <function>, arg1, arg2, etc...
                                        # call the function and tell it that 
                                        # the object in memory goes to the parameter 'data'
    .pipe((sns.lineplot, 'data'),hue='name',x='year',y='count')
) #everything in the middle of the parenthesis is one line of code
```
